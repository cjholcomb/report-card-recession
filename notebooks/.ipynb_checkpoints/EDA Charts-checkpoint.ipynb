```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
from scipy.interpolate import make_interp_spline, BSpline
sns.set_style('darkgrid')
import plotly.express as px
import json
from urllib.request import urlopen
with urlopen('https://raw.githubusercontent.com/plotly/datasets/master/geojson-counties-fips.json') as response:
    counties = json.load(response)




```

```python
df = pd.read_json('../data/training_dataset.json')
```

```python
df = df[df['10'] != 0]
df['delta_scaled'] = df['delta'] / df['10']
```

```python
df['recovery'].value_counts()
```

```python
df_2001 = df[df['year'] == 2001]
df_2008 = df[df['year'] == 2008]
```

```python
fig, ax = plt.subplots()
ax = sns.distplot(df_2001['delta_scaled'], kde = False, bins = 100, color = 'blue', label = '2001' )
ax = sns.distplot(df_2008['delta_scaled'], kde = False, bins = 100, color = 'red', label = '2008')
ax.get_yaxis().set_major_formatter(mtick.FuncFormatter(lambda x, p: format(int(x), ',')))
ax.set_title('Job Growth Distribution')
ax.set_ylabel('Occurance')
ax.set_xlabel('Job Growth Since 2001')
ax.set_xlim(-2, 2)
plt.legend();
```

```python
fig, ax = plt.subplots()
ax = sns.distplot(df_2001['decline'], kde = True, bins = 30, color = 'blue', label = '2001' )
ax = sns.distplot(df_2008['decline'], kde = True, bins = 50, color = 'red', label = '2008')
ax.axvline(x = df_2001['decline'].mean(), color = 'navy', ls= ':', linewidth = 0.5, alpha = 1, label = '2001 mean')
ax.axvline(x = df_2008['decline'].mean(), color = 'crimson', ls= ':', linewidth = 0.5, alpha = 1, label = '2001 mean')
ax.get_yaxis().set_major_formatter(mtick.FuncFormatter(lambda x, p: format(int(x), ',')))
ax.set_title('Time until recovery begins')
ax.set_ylabel('Occurance')
ax.set_xlabel('Quarters of Decline')
plt.legend();
```

```python
fig, ax = plt.subplots()
ax = sns.scatterplot(x = df['10'], y=df_2001['delta_scaled'], style = df['year'], size = df['decline'],  cmap ='RdBu')
ax.get_xaxis().set_major_formatter(mtick.FuncFormatter(lambda x, p: format(int(x), ',')))
ax.set_xlim(0, 100000)
ax.set_ylim(-2.5, 2.5)
ax.set_title('Job Growth X Total Jobs')
ax.set_ylabel('Job Growth')
ax.set_xlabel('Total Jobs')
plt.legend();
```

```python
fig = px.choropleth(df_2001, geojson = counties, locations='area_fips', color='delta_scaled', 
                           color_continuous_scale="piyg",
                           color_continuous_midpoint=0,
                           range_color = (-2,2),
                           scope="usa",
                           labels={'delta_scaled':'Job Growth'},
                           title= '2001 Job Growth'
                          )
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
# fig.suptitle('2001 Job Growth per capita')
plt.show()
```
